Import libaries

```
import tensorflow as tf
import numpy as ny
from tensorflow import keras
```

Prepare Data

```
xs = ny.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = ny.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

Create Model

```
layer = keras.layers.Dense(units=2,input_shape=[1])
layer1=keras.layers.Dense(units=2)
layer2=keras.layers.Dense(units=1)
model = tf.keras.Sequential([layer,layer1,layer2])
```

Apply optimizer and loss function

```
model.compile(optimizer='sgd',loss='mean_squared_error')
```

Train Model using fit function

```
model.fit(xs,ys,epochs=500)
```

Print weights and biases

```
print("Layer variables: {}".format(layer.get_weights()))
```

Predict unknown value

```
predicted_value = model.predict([10])
print(predicted_value)
```
